```python
import gradio as gr

def generate_html(user_input):

    html_code = f"""
<!DOCTYPE html>
<html>
<head>
<title>AI Generated Website</title>

<style>

body {{
font-family: Arial;
margin:0;
background:#f4f4f4;
}}

header {{
background:#4CAF50;
color:white;
padding:20px;
text-align:center;
}}

section {{
padding:20px;
margin:20px;
background:white;
border-radius:10px;
}}

footer {{
background:#333;
color:white;
text-align:center;
padding:10px;
}}

</style>

</head>

<body>

<header>
<h1>My AI Generated Website</h1>
</header>

<section>
<h2>Website Idea</h2>
<p>{user_input}</p>
</section>

<section>
<h2>About</h2>
<p>This webpage is automatically generated using AI HTML Quick Styler.</p>
</section>

<footer>
<p>Generated using AI</p>
</footer>

</body>
</html>
"""

    return html_code


interface = gr.Interface(
    fn=generate_html,
    inputs="text",
    outputs="code",
    title="AI HTML Page Generator"
)

interface.launch()
```
